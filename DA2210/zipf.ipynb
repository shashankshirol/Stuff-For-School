```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pop_data = pd.read_csv("data.csv")
```

```python
pop_data.describe()
```

```python
pop_data.columns
```

```python
pop_data_wo_country = pop_data.drop(columns=["Country"])
```

```python
pop_data_wo_country.head()
```

```python
pop_data_wo_country = pop_data_wo_country.rename(columns={"Frequency":"Frequency (in millions)"})
```

```python
pop_data_wo_country.plot(x="Rank", y="Frequency (in millions)", ylabel="Frequency", title="Frequency v Rank")
```

```python
pop_data_wo_country.plot(x="Rank", y="Frequency (in millions)", ylabel="log(Frequency)", xlabel="log(Rank)", loglog=True, kind="scatter", title="log(Frequency) v log(Rank)")
```

```python
log_rank = np.log10(pop_data_wo_country["Rank"])
log_freq = np.log10(pop_data_wo_country["Frequency (in millions)"])
```

```python
import seaborn as sns; sns.set_theme(color_codes=True)
ax = sns.regplot(x=log_rank, y=log_freq)
```

```python
a, b = np.polyfit(log_rank, log_freq, 1)
```
